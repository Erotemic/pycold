```python
import seaborn as sns
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pysccd
```

```python
t_c = -200

def getcategory(ccd_plot, i_curve):
    if ccd_plot[i_curve]['magnitude'][0, 3] > t_c and ccd_plot[i_curve]['magnitude'][0, 2] < -t_c and ccd_plot[i_curve]['magnitude'][0, 4] < -t_c:
        if ccd_plot[i_curve + 1]['coef'][3, 1] > abs(ccd_plot[i_curve]['coef'][3, 1]) and ccd_plot[i_curve + 1]['coef'][2, 1] \
             < -abs(ccd_plot[i_curve]['coef'][2, 1]) and ccd_plot[i_curve + 1]['coef'][4, 1] < -abs(ccd_plot[i_curve]['coef'][4, 1]):
            return 3
        else:
            return 2
    else:
        return 1
    
def pixel_plot_spcold(row, col, tile_h, tile_v, spcoldresults_path, in_path, out_path, band):
    # define fixed variables for ccd
   # ref_file = [name for name in os.listdir(in_path) if name.startswith('LC')][0]
    # tile_h = ref_file[3:6]
    # tile_v = ref_file[6:9]
    mode = 1
    method = 6
    probability = 0.99
    min_days_conse = 80
    output_mode = 11
    bandselectiobandit = 62
    user_mask_path = ''

    print('setting matplotlib..')
    sns.set(style="darkgrid")
    sns.set_context("notebook")
    f, [ax1, ax2] = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
    plt.subplots_adjust(left=0.08, right=0.98, top=0.92, bottom=0.1)
    labelfont = 12
    bias = 366

    # running pixel-based spcold
    print('processing spcold..')
    ccd_dt = np.dtype([('t_start', np.int32), ('t_end', np.int32), ('t_break', np.int32), ('pos', np.int32),
                       ('num_obs', np.int32), ('category', np.int32), ('t_confirmed', np.int32),
                       ('change_prob', np.int32),
                       ('coef', np.double, (7, 8)), ('rmse', np.double, (1, 7)), ('magnitude', np.double, (1, 7))])

    ret = pysccd.py_sccd(mode, in_path.encode('ascii'), out_path.encode('ascii'),
                         row, col, method, bandselectiobandit, spcoldresults_path.encode('ascii'), output_mode=output_mode,
                         probability_threshold=probability, min_days_conse=min_days_conse)
    spcold_result = np.fromfile(out_path + '/record_change_' + str(row) + '_' + str(col) + '_spcold.dat', dtype=ccd_dt)

    # running pixel-based cold
    print('processing cold..')
    method = 1
    ret = pysccd.py_sccd(mode, in_path.encode('ascii'), out_path.encode('ascii'),
                         row, col, method, bandselectiobandit, user_mask_path.encode('ascii'), output_mode=output_mode,
                         probability_threshold=probability, min_days_conse=min_days_conse)


    cold_result = np.fromfile(out_path + '/record_change_' + str(row) + '_' + str(col) + '_ccd.dat', dtype=ccd_dt)

    if band == 1:
        bandname = 'Blue'
        bn_shortname = '_B1'
    elif band == 2:
        bandname = 'Green'
        bn_shortname = '_B2'
    elif band == 3:
        bandname = 'Red'
        bn_shortname = '_B3'
    elif band == 4:
        bandname = 'NIR'
        bn_shortname = '_B4'
    elif band == 5:
        bandname = 'SWIR1'
        bn_shortname = '_B5'
    elif band == 6:
        bandname = 'SWIR2'
        bn_shortname = '_B6'
    elif band == 7:
        bandname = 'BT'
        bn_shortname = '_B7'


    # read obs records
    plot_obs = pd.read_csv(out_path + '/spectral_' + str(row) + '_' + str(col) + '_obs.csv')
    plot_obs.columns = ['Dates', 'Blue', 'Green', 'Red', 'NIR', \
                        'SWIR1', 'SWIR2', 'BT', 'QA', 'sensor']
    plot_obs_clean = plot_obs.loc[plot_obs['QA'] == 0]
    ordinaldates = [pd.Timestamp.fromordinal(row - bias)
                        for row in plot_obs_clean["Dates"]]
    plot_obs_clean.loc[:, 'Dates'] = ordinaldates

    #####################################
    # make spcold graph
    ###################################
    w = np.pi * 2 / 365.25
    # plot_state2 = pd.read_csv(out_path + '/StateRecords_' + str(row) + '_' + str(col) + '_' + bn_shortname + '.csv', na_values=-9999)
    plot_spcold = pd.DataFrame(columns=['Trend', 'Annual', 'Semiannual', 'Trimodel', 'predicted'])

    # plot actual observatyion
    ax1.plot('Dates', bandname, 'g+', data=plot_obs_clean,
             label='Observations')

    for i in range(len(spcold_result)):
        j = np.arange(spcold_result[i]['t_start'], spcold_result[i]['t_end'] + 1, 1)
        plot_spcold_sub = pd.DataFrame({'Dates': j,
                                     'Trend': j * spcold_result['coef'][i][band - 1][1] +
                                              spcold_result['coef'][i][band - 1][0],
                                     'Annual': np.cos(w * j) * spcold_result['coef'][i][band - 1][2] +
                                               np.sin(w * j) * spcold_result['coef'][i][band - 1][3],
                                     'Semiannual': np.cos(2 * w * j) * spcold_result['coef'][i][band - 1][4] +
                                                   np.sin(2 * w * j) * spcold_result['coef'][i][band - 1][5],
                                     'Trimodel': np.cos(3 * w * j) * spcold_result['coef'][i][band - 1][6] +
                                                 np.sin(3 * w * j) * spcold_result['coef'][i][band - 1][7]})

        plot_spcold_sub['predicted'] = plot_spcold_sub['Trend'] + plot_spcold_sub['Annual'] + plot_spcold_sub['Semiannual'] + \
                                        plot_spcold_sub['Trimodel']
        ordinaldates = [pd.Timestamp.fromordinal(int(row) - bias) for row in plot_spcold_sub["Dates"]]
        plot_spcold_sub.loc[:, 'Dates'] = ordinaldates
        g = sns.lineplot(x="Dates", y="predicted", data=plot_spcold_sub,
                         label="Model fit", ax=ax1,
                         color="orange")
        g.legend_.remove()
        plot_spcold = pd.concat([plot_spcold, plot_spcold_sub], sort=True)
    # set graph legend, title, etc

    ax1.set_ylabel("{} * 10000".format(bandname), fontsize=labelfont)
    ax1.set_xlabel("Dates", fontsize=labelfont)
    for tick in ax1.xaxis.get_major_ticks():
        tick.label.set_fontsize(12.5)
    ax1.annotate('A. SPCOLD', xy=(0.01, 0.85), xycoords='axes fraction', fontweight='bold', fontsize=18)

    for i in range(len(spcold_result) - 1):
        if getcategory(spcold_result, i) == 2:
            ax1.axvline(pd.Timestamp.fromordinal(spcold_result[i]['t_break'] - bias), color='r')
        else:
            ax1.axvline(pd.Timestamp.fromordinal(spcold_result[i]['t_break'] - bias), color='k')

    #####################################
    # make cold graph
    ###################################
    w = np.pi * 2 / 365.25
    # plot_state2 = pd.read_csv(out_path + '/StateRecords_' + str(row) + '_' + str(col) + '_' + bn_shortname + '.csv', na_values=-9999)
    plot_cold = pd.DataFrame(columns=['Trend', 'Annual', 'Semiannual', 'Trimodel', 'predicted'])

    # plot actual observatyion
    ax2.plot('Dates', bandname, 'g+', data=plot_obs_clean,
             label='Observations')

    for i in range(len(cold_result)):
        j = np.arange(cold_result[i]['t_start'], cold_result[i]['t_end'] + 1, 1)
        plot_cold_sub = pd.DataFrame({'Dates': j,
                                     'Trend': j * cold_result['coef'][i][band - 1][1] + cold_result['coef'][i][band - 1][0],
                                     'Annual': np.cos(w * j) * cold_result['coef'][i][band - 1][2] +
                                               np.sin(w * j) * cold_result['coef'][i][band - 1][3],
                                     'Semiannual': np.cos(2 * w * j) * cold_result['coef'][i][band - 1][4] +
                                                   np.sin(2 * w * j) * cold_result['coef'][i][band - 1][5],
                                     'Trimodel': np.cos(3 * w * j) * cold_result['coef'][i][band - 1][6] +
                                                 np.sin(3 * w * j) * cold_result['coef'][i][band - 1][7]})

        plot_cold_sub['predicted'] = plot_cold_sub['Trend'] + plot_cold_sub['Annual'] + plot_cold_sub['Semiannual'] + \
                                    plot_cold_sub['Trimodel']
        ordinaldates = [pd.Timestamp.fromordinal(int(row) - bias) for row in plot_cold_sub["Dates"]]
        plot_cold_sub.loc[:, 'Dates'] = ordinaldates
        g = sns.lineplot(x="Dates", y="predicted", data=plot_cold_sub,label="Model fit", ax=ax2, color="orange")
        g.legend_.remove()
        plot_cold = pd.concat([plot_cold, plot_cold_sub], sort=True)

    ax2.set_ylabel("{} * 10000".format(bandname), fontsize=labelfont)
    # set graph legend, title, etc

    ax1.set_ylabel("{} * 10000".format(bandname), fontsize=labelfont)
    ax1.set_xlabel("Dates", fontsize=labelfont)
    for tick in ax1.xaxis.get_major_ticks():
        tick.label.set_fontsize(12.5)
    ax2.annotate('B. COLD', xy=(0.01, 0.85), xycoords='axes fraction', fontweight='bold', fontsize=18)

    ax1.set_title('TILE H{}_V{}: Row_{} Col_{}'.format(tile_h, tile_v, row, col), size="x-large", fontweight="bold")
    # ax2.set_title("B. SCCD" + '_row' + str(row) + '_col' + str(col), size="x-large", fontweight="bold")

    for i in range(len(cold_result) - 1):
        if getcategory(cold_result, i) == 2:
            ax2.axvline(pd.Timestamp.fromordinal(cold_result[i]['t_break'] - bias), color='r')
        else:
            ax2.axvline(pd.Timestamp.fromordinal(cold_result[i]['t_break'] - bias), color='k')

    # finally generating legend
    from matplotlib.lines import Line2D
    legend_elements = [Line2D([0], [0], color='black', lw=1, label='Disturbance break'),
                       Line2D([0], [0], color='red', lw=1, label='Recovery break'),
                       Line2D([0], [0], color='orange', lw=2, label='Model fit'),
                       Line2D([], [], color='g', marker='+', linestyle='None',
                              markersize=8, label='observations')]
    ax1.legend(handles=legend_elements, loc='lower right', prop={'size': 7})

    # set a consistent plot ylimit for both axis
    extra = (np.max(plot_spcold['predicted']) - np.min(plot_spcold['predicted']))/4
    ax1.set(ylim=(np.min(plot_spcold['predicted'])- 1.5 * extra, np.max(plot_spcold['predicted'])+ 1.5 * extra))
    ax2.set(ylim=(np.min(plot_spcold['predicted'])- 1.5 * extra, np.max(plot_spcold['predicted'])+ 1.5 * extra))
    return spcold_result, cold_result

def pixel_plot_fixdays(row, col, tile_h, tile_v, in_path, out_path, band):
    # define fixed variables for ccd
   # ref_file = [name for name in os.listdir(in_path) if name.startswith('LC')][0]
    # tile_h = ref_file[3:6]
    # tile_v = ref_file[6:9]
    mode = 1
    method = 4
    probability = 0.99
    min_days_conse = 80
    output_mode = 11
    bandselectiobandit = 62
    user_mask_path = ''

    print('setting matplotlib..')
    sns.set(style="darkgrid")
    sns.set_context("notebook")
    f, [ax1, ax2] = plt.subplots(2, 1, figsize=(12, 6), sharex=True)
    plt.subplots_adjust(left=0.08, right=0.98, top=0.92, bottom=0.1)
    labelfont = 12
    bias = 366

    # running pixel-based fixdays cold
    print('processing fixdays cold..')
    ccd_dt = np.dtype([('t_start', np.int32), ('t_end', np.int32), ('t_break', np.int32), ('pos', np.int32),
                       ('num_obs', np.int32), ('category', np.int32), ('t_confirmed', np.int32),
                       ('change_prob', np.int32),
                       ('coef', np.double, (7, 8)), ('rmse', np.double, (1, 7)), ('magnitude', np.double, (1, 7))])

    ret = pysccd.py_sccd(mode, in_path.encode('ascii'), out_path.encode('ascii'),
                         row, col, method, bandselectiobandit, user_mask_path.encode('ascii'), output_mode=output_mode,
                         probability_threshold=probability, min_days_conse=min_days_conse)
    fixdays_result = np.fromfile(out_path + '/record_change_' + str(row) + '_' + str(col) + '_ccd.dat', dtype=ccd_dt)
    os.remove(out_path + '/record_change_' + str(row) + '_' + str(col) + '_ccd.dat')
    
    # running pixel-based cold
    print('processing cold..')
    method = 1
    ret = pysccd.py_sccd(mode, in_path.encode('ascii'), out_path.encode('ascii'),
                         row, col, method, bandselectiobandit, user_mask_path.encode('ascii'), output_mode=output_mode,
                         probability_threshold=probability, min_days_conse=min_days_conse)


    cold_result = np.fromfile(out_path + '/record_change_' + str(row) + '_' + str(col) + '_ccd.dat', dtype=ccd_dt)

    if band == 1:
        bandname = 'Blue'
        bn_shortname = '_B1'
    elif band == 2:
        bandname = 'Green'
        bn_shortname = '_B2'
    elif band == 3:
        bandname = 'Red'
        bn_shortname = '_B3'
    elif band == 4:
        bandname = 'NIR'
        bn_shortname = '_B4'
    elif band == 5:
        bandname = 'SWIR1'
        bn_shortname = '_B5'
    elif band == 6:
        bandname = 'SWIR2'
        bn_shortname = '_B6'
    elif band == 7:
        bandname = 'BT'
        bn_shortname = '_B7'


    # read obs records
    plot_obs = pd.read_csv(out_path + '/spectral_' + str(row) + '_' + str(col) + '_obs.csv')
    plot_obs.columns = ['Dates', 'Blue', 'Green', 'Red', 'NIR', \
                        'SWIR1', 'SWIR2', 'BT', 'QA', 'sensor']
    plot_obs_clean = plot_obs.loc[plot_obs['QA'] == 0]
    ordinaldates = [pd.Timestamp.fromordinal(row - bias)
                        for row in plot_obs_clean["Dates"]]
    plot_obs_clean.loc[:, 'Dates'] = ordinaldates

    #####################################
    # make fixedays graph
    ###################################
    w = np.pi * 2 / 365.25
    # plot_state2 = pd.read_csv(out_path + '/StateRecords_' + str(row) + '_' + str(col) + '_' + bn_shortname + '.csv', na_values=-9999)
    plot_fixdays = pd.DataFrame(columns=['Trend', 'Annual', 'Semiannual', 'Trimodel', 'predicted'])

    # plot actual observatyion
    ax1.plot('Dates', bandname, 'g+', data=plot_obs_clean,
             label='Observations')

    for i in range(len(fixdays_result)):
        j = np.arange(fixdays_result[i]['t_start'], fixdays_result[i]['t_end'] + 1, 1)
        plot_fixdays_sub = pd.DataFrame({'Dates': j,
                                     'Trend': j * fixdays_result['coef'][i][band - 1][1] +
                                              fixdays_result['coef'][i][band - 1][0],
                                     'Annual': np.cos(w * j) * fixdays_result['coef'][i][band - 1][2] +
                                               np.sin(w * j) * fixdays_result['coef'][i][band - 1][3],
                                     'Semiannual': np.cos(2 * w * j) * fixdays_result['coef'][i][band - 1][4] +
                                                   np.sin(2 * w * j) * fixdays_result['coef'][i][band - 1][5],
                                     'Trimodel': np.cos(3 * w * j) * fixdays_result['coef'][i][band - 1][6] +
                                                 np.sin(3 * w * j) * fixdays_result['coef'][i][band - 1][7]})

        plot_fixdays_sub['predicted'] = plot_fixdays_sub['Trend'] + plot_fixdays_sub['Annual'] + plot_fixdays_sub['Semiannual'] + \
                                        plot_fixdays_sub['Trimodel']
        ordinaldates = [pd.Timestamp.fromordinal(int(row) - bias) for row in plot_fixdays_sub["Dates"]]
        plot_fixdays_sub.loc[:, 'Dates'] = ordinaldates
        g = sns.lineplot(x="Dates", y="predicted", data=plot_fixdays_sub,
                         label="Model fit", ax=ax1,
                         color="orange")
        g.legend_.remove()
        plot_fixdays = pd.concat([plot_fixdays, plot_fixdays_sub], sort=True)
    # set graph legend, title, etc

    ax1.set_ylabel("{} * 10000".format(bandname), fontsize=labelfont)
    ax1.set_xlabel("Dates", fontsize=labelfont)
    for tick in ax1.xaxis.get_major_ticks():
        tick.label.set_fontsize(12.5)
    ax1.annotate('A. fixdays COLD', xy=(0.01, 0.85), xycoords='axes fraction', fontweight='bold', fontsize=18)

    for i in range(len(fixdays_result) - 1):
        if getcategory(fixdays_result, i) == 2:
            ax1.axvline(pd.Timestamp.fromordinal(fixdays_result[i]['t_break'] - bias), color='r')
        else:
            ax1.axvline(pd.Timestamp.fromordinal(fixdays_result[i]['t_break'] - bias), color='k')

    #####################################
    # make cold graph
    ###################################
    w = np.pi * 2 / 365.25
    # plot_state2 = pd.read_csv(out_path + '/StateRecords_' + str(row) + '_' + str(col) + '_' + bn_shortname + '.csv', na_values=-9999)
    plot_cold = pd.DataFrame(columns=['Trend', 'Annual', 'Semiannual', 'Trimodel', 'predicted'])

    # plot actual observatyion
    ax2.plot('Dates', bandname, 'g+', data=plot_obs_clean,
             label='Observations')

    for i in range(len(cold_result)):
        j = np.arange(cold_result[i]['t_start'], cold_result[i]['t_end'] + 1, 1)
        plot_cold_sub = pd.DataFrame({'Dates': j,
                                     'Trend': j * cold_result['coef'][i][band - 1][1] + cold_result['coef'][i][band - 1][0],
                                     'Annual': np.cos(w * j) * cold_result['coef'][i][band - 1][2] +
                                               np.sin(w * j) * cold_result['coef'][i][band - 1][3],
                                     'Semiannual': np.cos(2 * w * j) * cold_result['coef'][i][band - 1][4] +
                                                   np.sin(2 * w * j) * cold_result['coef'][i][band - 1][5],
                                     'Trimodel': np.cos(3 * w * j) * cold_result['coef'][i][band - 1][6] +
                                                 np.sin(3 * w * j) * cold_result['coef'][i][band - 1][7]})

        plot_cold_sub['predicted'] = plot_cold_sub['Trend'] + plot_cold_sub['Annual'] + plot_cold_sub['Semiannual'] + \
                                    plot_cold_sub['Trimodel']
        ordinaldates = [pd.Timestamp.fromordinal(int(row) - bias) for row in plot_cold_sub["Dates"]]
        plot_cold_sub.loc[:, 'Dates'] = ordinaldates
        g = sns.lineplot(x="Dates", y="predicted", data=plot_cold_sub,label="Model fit", ax=ax2, color="orange")
        g.legend_.remove()
        plot_cold = pd.concat([plot_cold, plot_cold_sub], sort=True)

    ax2.set_ylabel("{} * 10000".format(bandname), fontsize=labelfont)
    # set graph legend, title, etc

    ax1.set_ylabel("{} * 10000".format(bandname), fontsize=labelfont)
    ax1.set_xlabel("Dates", fontsize=labelfont)
    for tick in ax1.xaxis.get_major_ticks():
        tick.label.set_fontsize(12.5)
    ax2.annotate('B. COLD', xy=(0.01, 0.85), xycoords='axes fraction', fontweight='bold', fontsize=18)

    ax1.set_title('TILE H{}_V{}: Row_{} Col_{}'.format(tile_h, tile_v, row, col), size="x-large", fontweight="bold")
    # ax2.set_title("B. SCCD" + '_row' + str(row) + '_col' + str(col), size="x-large", fontweight="bold")

    for i in range(len(cold_result) - 1):
        if getcategory(cold_result, i) == 2:
            ax2.axvline(pd.Timestamp.fromordinal(cold_result[i]['t_break'] - bias), color='r')
        else:
            ax2.axvline(pd.Timestamp.fromordinal(cold_result[i]['t_break'] - bias), color='k')

    # finally generating legend
    from matplotlib.lines import Line2D
    legend_elements = [Line2D([0], [0], color='black', lw=1, label='Disturbance break'),
                       Line2D([0], [0], color='red', lw=1, label='Recovery break'),
                       Line2D([0], [0], color='orange', lw=2, label='Model fit'),
                       Line2D([], [], color='g', marker='+', linestyle='None',
                              markersize=8, label='observations')]
    ax1.legend(handles=legend_elements, loc='lower right', prop={'size': 7})

    # set a consistent plot ylimit for both axis
    extra = (np.max(plot_fixdays['predicted']) - np.min(plot_fixdays['predicted']))/4
    ax1.set(ylim=(np.min(plot_fixdays['predicted'])- 1.5 * extra, np.max(plot_fixdays['predicted'])+ 1.5 * extra))
    ax2.set(ylim=(np.min(plot_fixdays['predicted'])- 1.5 * extra, np.max(plot_fixdays['predicted'])+ 1.5 * extra))
    return fixdays_result, cold_result
```
